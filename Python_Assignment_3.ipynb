```
def reduce(op , arr) :
  var = arr[0]
  for i in arr[1 : ] :
    var = op(var , i) 
  return var
```

```
array = [1 , 2 , 4 , 1 , 50 , 27 , 100 , 12]
print('Array - ' + str(array))
def op(a , b) : return max(a , b)
print('Output - ' +  str(reduce(op , array)))
```

```
def filter(op , arr) :
  temp = []
  for i in arr :
    if op(i) :
      temp.append(i)
  return temp
```

```
array = [1 , 2 , 4 , 1 , 50 , 27 , 100 , 12]
print('Array - ' + str(array))
def op(a) : return a % 2 == 0
print('Output - ' +  str(filter(op , array)))
```
