```
class InputTriangle :
    def __init__(self) :
        self.s1 = float(input('Side 1 : '))
        self.s2 = float(input('Side 2 : '))
        self.s3 = float(input('Side 3 : '))
class GetArea(InputTriangle) :
    def calc(self) :
        s = (self.s1 + self.s2 + self.s3) / 2
        area = (s * (s - self.s1) * (s - self.s2) * (s - self.s3)) ** 0.5
        print('Area : ' + str(area) + ' cm\u00b2')
```

```
area = GetArea()
area.calc()
```

```
def filter_long_words(words , n) :
    ans = []
    for i in words :
        if len(i) > n :
            ans.append(i)
    return ans
```

```
words = ['red' , 'yellow' , 'blue' , 'orange' , 'green' , 'purple']
n = 5
print('Output : ' + str(filter_long_words(words , n)))
```

```
def word_length(word) : return len(word)

words = ['red' , 'yellow' , 'blue' , 'orange' , 'green' , 'purple']
print('Output : ' + str(list(map(word_length , words))))
```

```
def vowel(char):
    if char == "a" or char == "e" or char == "i" or char == "o" or char == "u":
        return True
    else:
        return False

vowel(input('Enter Character - '))
```
